```python
import pandas as pd
```

```python
# Load the dataset
file_path = r'E:\epic\Training_Data_Google_Play_reviews_6000.xlsx'
data = pd.read_excel(file_path)
```

```python
# Display the first few rows of the dataset
data.head(), data.info()
```

```python
# Check for missing values in the dataset
missing_values = data.isnull().sum()
missing_values
```

```python
# Drop rows with missing content
data_cleaned = data.dropna(subset=['content'])
```

```python
# Drop columns with too many missing values
data_cleaned = data_cleaned.drop(columns=['replyContent', 'repliedAt'])
```

```python

# Display the cleaned dataset information
data_cleaned.info(), data_cleaned.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
```

```python
# Set up the visualization style
sns.set(style="whitegrid")
```

```python
# Distribution of ratings
plt.figure(figsize=(10, 6))
sns.countplot(x='score', data=data_cleaned, palette='viridis')
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

```python
# Relationship between ratings and thumbs-up count
plt.figure(figsize=(10, 6))
sns.boxplot(x='score', y='thumbsUpCount', data=data_cleaned, palette='viridis')
plt.yscale('log')  # Use log scale for better visualization of outliers
plt.title('Relationship Between Ratings and Thumbs-Up Count')
plt.xlabel('Rating')
plt.ylabel('Thumbs-Up Count (log scale)')
plt.show()
```
